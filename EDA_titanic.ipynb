```python
survival_rates = df.groupby("Sex")["Survived"].mean()

plt.bar(survival_rates.index, survival_rates.values, color=["pink", "skyblue"])
plt.title("Survival Rate by Gender")
plt.ylabel("Survival Probability")
plt.xlabel("Gender")
plt.show()
```

```python
urvival_rates = df.groupby("Embarked")["Survived"].mean()


plt.pie(
    urvival_rates.values, 
    labels=urvival_rates.index, 
    autopct="%1.1f%%",  
    startangle=90,      
    colors=["lightgreen", "skyblue", "salmon"]
)

plt.title("Survival Rate by Embarkation Point")
plt.show()
```

-------------------------------------------------------------------------------------------------------------------------------

Insight: From the Titanic dataset analysis, it can be observed that certain groups of passengers had a much higher probability of survival. In particular, a passenger who embarked from Cherbourg (France), was female, belonged to the 1st class, and was assigned to Deck B, D, or E had significantly better chances of survival. This likelihood increased further if the passenger was near her 20s and had a small family size (not traveling alone but also not in a very large group). These factors collectively indicate that survival on the Titanic was strongly influenced by a combination of gender, class, deck location, age, embarkation point, and family size, with females from wealthier backgrounds having the highest chances of survival.
